```python
print("Displaying odd numbers from 1 to 20:")

for num in range(1, 21):
    if num % 2 == 0:
        continue

    print(num)
```

```python
# Password Checker
secret_password = "24446666"
max_attempts = 3
attempts = 0

while True:
    user_input = input("Enter the password: ")
    attempts += 1

    if user_input == secret_password:
        print("Access granted!")
        break
    else:
        if attempts < max_attempts:
            print(f"Incorrect. You have {max_attempts - attempts} attempts left.")
        else:
            print("Access Denied. Too many incorrect attempts.")
            break
```

```python
# Number Filter Display
print("Filtering numbers from 1 to 10:")

for i in range(1, 11):
    if i == 8:
        print("Reached 8. Stopping loop.")
        break

    if i == 4:
        continue

    if i == 6:
        pass  # pass acts as a placeholder and does nothing

    print(i)
```
